```python
import pandas as pd

# What's in DataFrame?
df = pd.read_csv('../data/auto-mpg.csv')

# Seeing the structure
print(type(df)) 
print(df.shape)
print(df.columns)
df.dtypes
df.describe()
df['mpg'] # Bracket notation for selecting a single column

# iloc not inclusive meaning doesn't include 5
df.iloc[2:5]

# loc this inclusive
df.loc[0:5, 'mpg':'weight']

# Boolean filtering
df[df['mpg'] > 30]

# Mixing Boolean Filtering with Loc
subset = df.loc[1:100]
subset[subset['mpg'] > 30]

# Boolean Filtering with Multiple Conditions
df[(df['mpg'] > 30) & (df['cylinders'] == 4)]

# Filter by string
df[df['name'].str.contains('ford')]
```

```python
import pandas as pd

df = pd.read_csv('../data/auto-mpg.csv', na_values=['NA', '?', 'NA',''])

# 1. How many missing values in each column?
df.isnull().sum()

# 2. Which rows have missing horsepower?
df[df['horsepower'].isnull()]

# 3. Fill missing horsepower with the median
median_hp = df['horsepower'].median()
df['horsepower'] = df['horsepower'].fillna(median_hp)

# 4. Verify no more missing values in horsepower
df[df['horsepower'].isnull()]
```
